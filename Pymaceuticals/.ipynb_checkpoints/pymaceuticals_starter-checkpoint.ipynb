## Observations and Insights

## Dependencies and starter code

```python
#Three data observations:
#1. The male to female ratio in mice is almost equal
#2. The data for Infubinol is the most evenly distributed
#3. We have the most data for Ramicane and Capomulin treatments


# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
combined_studydata = pd.merge(mouse_metadata, study_results, how="right",on="Mouse ID")
combined_studydata.head(5)
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

best_treatment = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

tumor_stats = combined_studydata.groupby(['Drug Regimen']).agg(["count","mean", "median","var","std","sem"])['Tumor Volume (mm3)']

tumor_stats_reset = pd.DataFrame(tumor_stats).reset_index()



tumor_stats_best = tumor_stats_reset[tumor_stats_reset['Drug Regimen'].isin(["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]) ]


tumor_stats_best
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 

best_treatment = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

combined_studydata_best = combined_studydata[combined_studydata['Drug Regimen'].isin(["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]) ]

combined_studydata_best

mouse_data = combined_studydata_best.loc[:,["Mouse ID", "Timepoint", "Tumor Volume (mm3)", "Drug Regimen", "Weight (g)"]]

max_timepoint = mouse_data.groupby(["Drug Regimen","Mouse ID"])["Timepoint"].max().reset_index()


max_timepoint_merge=pd.merge(max_timepoint, mouse_data, on=["Mouse ID","Timepoint"] , how="left")
del max_timepoint_merge['Drug Regimen_y']
max_timepoint_merge.rename(columns={'Drug Regimen_x': 'Drug Regimen'}, inplace=True)


Tumor_Volume = max_timepoint_merge["Tumor Volume (mm3)"]


quartiles = Tumor_Volume.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Tumor Volume is: {lowerq}")
print(f"The upper quartile of Tumor Volume is: {upperq}")
print(f"The interquartile range of Tumor Volume is: {iqr}")
print(f"The the median of Tumor Volume is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
print(max_timepoint_merge)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

bplot=sns.boxplot(y='Tumor Volume (mm3)', x='Drug Regimen', 
                 data=max_timepoint_merge, 
                 width=0.5,
                 palette="colorblind")

bplot.axes.set_title("Final Tumor Volume",
                    fontsize=16)
bplot
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin


capomulin_studydata = combined_studydata[combined_studydata['Drug Regimen'].isin(["Capomulin"])]



capomulin_studydata_pivot= capomulin_studydata.pivot(index='Timepoint', columns='Mouse ID', values='Tumor Volume (mm3)')
    
capomulin_studydata_pivot.plot()

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title("Timepoint vs Tumor Volume per Mouse")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
average_tumor_cap = capomulin_studydata.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()


avg_tum = pd.DataFrame({"Average Tumor Volume": average_tumor_cap})

scatter_data=pd.merge(avg_tum, capomulin_studydata, on=["Mouse ID"] , how="left")

scatter_data.plot.scatter(x="Weight (g)", y="Average Tumor Volume", title= "Scatter plot Weight vs Volume Capomulin")

```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen

weight = scatter_data.iloc[:,5]
avg_tumor = scatter_data.iloc[:,1]
correlation = st.pearsonr(weight,avg_tumor)

print(f"The correlation between both factors is {round(correlation[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(weight, avg_tumor)
#vc_slope, vc_int, vc_r, vc_p, vc_std_err = st.linregress(weight, avg_tumor)
print(f"The slope is {(slope)}")
print(f"The intercept is {intercept}")
print(f"The R-Value is {rvalue}")
print(f"The P-Value is {pvalue}")
print(f"The standard error is {stderr}")
#Plot the linear regression model on top of the previous scatter plot.

regress_values = weight * slope + intercept

plt.scatter(weight,avg_tumor)
plt.plot(weight,regress_values,"r-")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(6,10),fontsize=26,color="red")
plt.title("Scatter plot Weight vs Volume Capomulin")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume')
plt.show()

```
